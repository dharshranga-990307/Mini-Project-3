```python

from scipy.stats import f_oneway



group1 = df3.groupby('city')['price'].apply(list)
F1, p1 = f_oneway(*group1)
print(f"Feature1 - ANOVA F-statistic: {F1}, p-value: {p1}")

# Perform ANOVA for Feature2
group2 = df3.groupby('ft')['price'].apply(list)
F2, p2 = f_oneway(*group2)
print(f"Feature2 - ANOVA F-statistic: {F2}, p-value: {p2}")

group3 = df3.groupby('bt')['price'].apply(list)
F3, p3 = f_oneway(*group3)
print(f"Feature3 - ANOVA F-statistic: {F3}, p-value: {p3}")

# Perform ANOVA for Feature2
group4 = df3.groupby('oem')['price'].apply(list)
F4, p4 = f_oneway(*group4)
print(f"Feature4 - ANOVA F-statistic: {F4}, p-value: {p4}")

group5 = df3.groupby('transmission')['price'].apply(list)
F5, p5 = f_oneway(*group5)
print(f"Feature5 - ANOVA F-statistic: {F5}, p-value: {p5}")

# Perform ANOVA for Feature2
group6 = df3.groupby('Insurance Validity')['price'].apply(list)
F6, p6 = f_oneway(*group6)
print(f"Feature6 - ANOVA F-statistic: {F6}, p-value: {p6}")

group7 = df3.groupby('Ownership')['price'].apply(list)
F7, p7 = f_oneway(*group7)
print(f"Feature7 - ANOVA F-statistic: {F7}, p-value: {p7}")

# Perform ANOVA for Feature2
group8 = df3.groupby('Color')['price'].apply(list)
F8, p8 = f_oneway(*group8)
print(f"Feature8 - ANOVA F-statistic: {F8}, p-value: {p8}")

group9 = df3.groupby('model')['price'].apply(list)
F9, p9 = f_oneway(*group9)
print(f"Feature9 - ANOVA F-statistic: {F9}, p-value: {p9}")


# Select features with significant p-values
selected_features = []
if p1 < 0.05:
    selected_features.append('Feature1')
if p2 < 0.05:
    selected_features.append('Feature2')
if p3 < 0.05:
    selected_features.append('Feature3')
if p4 < 0.05:
    selected_features.append('Feature4')
if p5 < 0.05:
    selected_features.append('Feature5')
if p6 < 0.05:
    selected_features.append('Feature6')
if p7 < 0.05:
    selected_features.append('Feature7')
if p8 < 0.05:
    selected_features.append('Feature8')
if p9 < 0.05:
    selected_features.append('Feature9')

print(f"Selected Features: {selected_features}")
```

```python
from scipy.stats import chi2_contingency

def chi_square_test(data, target):
    results = []
    for column in data.columns:
        if data[column].dtype == 'object':
            contingency_table = pd.crosstab(data[column], target)
            chi2, p, dof, expected  = chi2_contingency(contingency_table)
            results.append((column, p))
    return [col for col, p in results if p < 0.05]

selected_columns = chi_square_test(df3.drop('price', axis=1), df3['price'])
print(selected_columns)
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm



# Plot histogram
plt.hist(df3['Engine Displacement'], bins=5, density=True, alpha=0.6, color='g')

# Plot the normal distribution curve
mean, std = df3['Engine Displacement'].mean(), df3['Engine Displacement'].std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)
title = " Engine Displacement-Fit results: mu = %.2f,  std = %.2f" % (mean, std)
plt.title(title)

plt.show()
```
